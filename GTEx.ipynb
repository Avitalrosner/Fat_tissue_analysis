```python
delete_genes = ["FAU", "MRPL13", "RPL10", "RPL10A", "RPL10L", "RPL11", "RPL12", "RPL13", "RPL13A", "RPL14",
    "RPL15", "RPL17", "RPL18", "RPL18A", "RPL19", "RPL21", "RPL22", "RPL22L1", "RPL23", "RPL23A",
    "RPL24", "RPL26", "RPL26L1", "RPL27", "RPL27A", "RPL28", "RPL29", "RPL3", "RPL30", "RPL31",
    "RPL32", "RPL34", "RPL35", "RPL35A", "RPL36", "RPL36A", "RPL36AL", "RPL37", "RPL37A", "RPL38",
    "RPL39", "RPL3L", "RPL4", "RPL41", "RPL5", "RPL6", "RPL7", "RPL7A", "RPL8", "RPL9", "RPLP0",
    "RPLP1", "RPLP2", "RPS10", "RPS11", "RPS12", "RPS13", "RPS15", "RPS15A", "RPS16", "RPS17",
    "RPS18", "RPS19", "RPS2", "RPS20", "RPS21", "RPS23", "RPS24", "RPS25", "RPS26", "RPS27",
    "RPS27A", "RPS27L", "RPS28", "RPS29", "RPS3", "RPS3A", "RPS4X", "RPS4Y1", "RPS5", "RPS6",
    "RPS7", "RPS8", "RPS9", "RPSA", "RSL24D1", "RSL24D1P11", "UBA52"]
df_no_mt_ribo = df_without_mt[~df_without_mt["Description"].isin(delete_genes)]
df_no_mt_ribo
```

```python
df_no_mt_ribo.set_index("Description", inplace=True)
df_no_mt_ribo
```

```python
df_filtered = df_no_mt_ribo.drop("Name", axis = 1)
```

```python
df_only_adipose_subcutaneous = df_filtered.drop(['Adipose - Visceral (Omentum)', 'Breast - Mammary Tissue'], axis = 1)
df_only_adipose_visceral = df_filtered.drop(['Adipose - Subcutaneous', 'Breast - Mammary Tissue'], axis = 1)
df_only_breast = df_filtered.drop(["Adipose - Visceral (Omentum)", 'Adipose - Subcutaneous'], axis = 1)
```

```python
def find_marker_genes(df):
    ln=1e-9             
    THRESHOLD = 2  
    marker_genes = dict()                       
    for my_tissue in df.columns:             
        my_tissue_expression = df[my_tissue]  
        other_tissues_max_expression = df[df.columns[df.columns!=my_tissue]].max(axis=1) 
        tissue_ratio = (my_tissue_expression+ln)/(other_tissues_max_expression+ln)  
        marker_genes[my_tissue] = tissue_ratio[tissue_ratio > THRESHOLD]
        return marker_genes

only_adipose_subcutaneous = find_marker_genes(df_only_adipose_subcutaneous)
only_adipose_visceral = find_marker_genes(df_only_adipose_visceral)
only_breast =  find_marker_genes(df_only_breast)
all_markers = find_marker_genes(df_filtered)

```

```python
vis = list(only_adipose_visceral.values())
sub = list(only_adipose_subcutaneous.values())
bre = list(only_breast.values())
```

```python
lst_vis = []
df_split_vis = pd.DataFrame(vis)
for gene in df_split_vis:
    lst_vis.append(gene)
```

```python
lst_sub = []
df_split_sub = pd.DataFrame(sub)
for gene in df_split_sub:
    lst_sub.append(gene)
```

```python
lst_bre = []
df_split_bre = pd.DataFrame(bre)
for gene in df_split_bre:
    lst_bre.append(gene)
```

```python
union_vis_sub = lst_sub + lst_vis         
union_vis_sub = list(set(union_vis_sub)) 
sorted_lst = sorted(union_vis_sub)
intersect_bre_union = list(set(union_vis_sub).intersection(set(lst_bre)))
intersect_bre_union
```

```python
# This function get normallized dataframe (each column is a sample, each row is a gene), and it returns the marker genes for each sample in a dictionary
# There is an option to plot the figure with all the datapoints to validate that the results makes sense
def get_marker_genes(df, FC_THRESH=2, MAX_EXP_THRESH = 1e-4, PN=1e-9, SHOW_FIG=False):
    # normalize for any case
    df = df/df.sum(axis=0) 

    if SHOW_FIG:
        n_side = int(np.ceil(len(df.columns)**0.5))
        fig = make_subplots(rows=n_side, cols=n_side, subplot_titles=df.columns, vertical_spacing=0.25)

    marker_genes_dict = dict()
    for ii,my_tissue in enumerate(df.columns):
        my_tissue_expression = df[my_tissue]
        other_tissues_max_expression = df[df.columns[df.columns!=my_tissue]].max(axis=1)
        all_tissues_max_exp = df.max(axis=1)
        tissue_ratio = (my_tissue_expression[all_tissues_max_exp>MAX_EXP_THRESH]+PN)/(other_tissues_max_expression[all_tissues_max_exp>MAX_EXP_THRESH]+PN)
        marker_genes_dict[my_tissue] = tissue_ratio[tissue_ratio > FC_THRESH]
    
        if SHOW_FIG:
            row_idx = int(np.ceil((ii+1-n_side)/n_side)+1)
            col_idx = int((ii%n_side)+1)
            fig.add_scatter(x=np.log10(my_tissue_expression+PN),y=np.log10(other_tissues_max_expression+PN), mode='markers', 
                            hovertext=my_tissue_expression.index, marker=dict(color='#0000FF'),
                row=row_idx,col=col_idx)
            
            fig.add_scatter(x=np.log10(my_tissue_expression[marker_genes_dict[my_tissue].index]+PN),y=np.log10(other_tissues_max_expression[marker_genes_dict[my_tissue].index]+PN), mode='markers', 
                            hovertext=my_tissue_expression[marker_genes_dict[my_tissue].index].index, marker=dict(color='#FF0000'),
                row=row_idx,col=col_idx)
            
            fig.update_yaxes(title_text="Other log10 max", row=row_idx, col=col_idx)
            fig.update_xaxes(title_text=my_tissue+" log10 exp", row=row_idx, col=col_idx)

    if SHOW_FIG:
        fig.update_layout(showlegend=False, title_text="Marker genes for each tissue (Min exp="+str(MAX_EXP_THRESH)+", FC thersh="+str(FC_THRESH)+")")
        fig.show()
    return marker_genes_dict

get_marker_genes(df_filtered)
   
```
